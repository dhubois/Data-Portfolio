```python
sns.stripplot(data=df, x='pulse')
```

```python
sns.ecdfplot(data=df, x='pulse')
```

#### Using the Grouping Variable (HUE)

```python
sns.ecdfplot(data=df, x='pulse', hue='diet')
```

### Simple Combination of Univariate Graphs

```python
sns.kdeplot(data=df, x='pulse', hue='diet')
sns.rugplot(data=df, x='pulse', hue='diet')
```

#### Plots With 2 Variables (Bivariate Analysis)

Possible **Bivariate Plots**

1. Continuous
    * Scatterplot
    * LMP plot
    * Reg plot
    * Hex plot
    * Pair Grid
    * KDE plot
    
2. Categorical
    * Line plot
    * Bar plot
    * Point plot

```python
#read in the data
df_fly = sns.load_dataset('flights')
df_fly.head()
```

```python
sns.lineplot(data=df_fly, x='year', y='passengers')
```

```python
sns.pointplot(data=df_fly, x='year', y='passengers')
```

```python
sns.barplot(data=df_fly, x='year', y='passengers')
```

```python
df_dia = sns.load_dataset('diamonds')
df_dia.head()
```

```python
df_dia = df_dia.sample(3000)
```

```python
sns.scatterplot(data=df_dia, x='carat', y='price')
```

```python
sns.scatterplot(data=df_dia, x='carat', y='price', marker='+')
```

```python
sns.kdeplot(data=df_dia, x='carat', y='price')
```

```python
sns.kdeplot(data=df_dia, x='carat', y='price', fill=True, cmap='mako_r')
```

```python
sns.histplot(data=df_dia, x='carat', y='price')
```

#### Combining two Bivariate plots (KDE + HistPlot)

```python
sns.kdeplot(data=df_dia, x='carat', y='price', cmap='coolwarm')
sns.histplot(data=df_dia, x='carat', y='price', color='red')
```

```python
sns.jointplot(data=df_dia, x='carat', y='price')
```

```python
sns.jointplot(data=df_dia, x='carat', y='price', kind='hex')
```

```python
sns.lmplot(data=df_dia, x='carat', y='price', markers='.', line_kws={'color':'red'})
```

### Plotting Relationships

```python
sns.relplot(data=df_dia, x='carat', y='price', col='cut')
```
